```python
import pandas as pd
import numpy as np
import random
import time
import math
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mockdata.csv")
df['basePrice'] = df.apply(lambda row: float(row['basePrice'].strip()[1:]), axis=1)
df['quantity'] = np.random.randint(1,50,df.shape[0])
df['wastage'] = np.random.randint(0,50, df.shape[0])/100
df['dynamicPrice'] = df.apply(lambda row: row['basePrice'] - 0.6*row['basePrice']/(1+math.log(row['daysToExpiry'])), axis=1)
df
```

```python
model = RandomForestRegressor(random_state=1, n_estimators=10)
```

```python
x = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
model.fit(x[100:], y[100:])
pred = model.predict(x)
```

```python
df['predictedPrice'] = pred
df
```
